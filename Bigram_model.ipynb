```python
import torch
import torch.optim as optim

# Assuming xenc, xs, and ys are defined somewhere in your code

g = torch.Generator().manual_seed(2147483647)
w = torch.randn((27, 27), generator=g, requires_grad=True)

optimizer = optim.SGD([w], lr=0.01)

for epoch in range(100):
    # Create a new loss tensor for each epoch
    loss_values = []

    logits = torch.matmul(xenc, w)
    logged = logits.exp()
    prob = logged / logged.sum(1, keepdim=True)

    for i in range(prob.size(0)):
        x = xs[i].item()
        y = ys[i].item()
        l_1 = -prob[i, y]
        loss_values.append(torch.log(l_1 + 1e-12))  # Added a small epsilon to avoid log(0)

        # if (i + 1) % 10 == 0:
        #     print(f'The loss at epoch {epoch + 1}, iteration {i + 1}: {loss_values[-1].item()}')

    # Create a new tensor to store the losses
    loss = torch.stack(loss_values)

    # Calculate the average loss for the epoch
    average_loss = loss.mean()

    # Backward pass and optimization step
    average_loss.backward()
    optimizer.step()
    optimizer.zero_grad()

    print(f'Epoch {epoch + 1}, Average loss: {average_loss.item()}')
```
